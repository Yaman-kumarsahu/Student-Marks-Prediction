```python
import pandas as pd
```

```python
ese_data = pd.read_excel('ese_dsa.xlsx')
ese_data
```

```python
sess_data = pd.read_excel('sess_dsa.xlsx')
sess_data
```

```python
temp = {'roll' : [], 'ct1' : [], 'ct2' : [], 'att' : [], 'ta' : [], 'ese' : []}
l = []
for i in range(108):
    for j in range(109):
        if sess_data['roll'][i] == ese_data['roll'][j]:
            temp['roll'].append(sess_data['roll'][i])
            temp['ct1'].append(sess_data['ct1'][i])
            temp['ct2'].append(sess_data['ct2'][i])
            temp['att'].append(sess_data['att'][i])
            temp['ta'].append(sess_data['ta'][i])
            temp['ese'].append(ese_data['ese'][j])
            l.append(sess_data['roll'][i])
```

```python
len(temp['ese'])
```

```python
final_data = pd.DataFrame(temp)
final_data
```

```python
final_data.to_excel('deld_final.xlsx')
```
